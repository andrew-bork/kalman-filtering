```python

import sympy
from sympy.abc import alpha, x, y, v, w, R, theta
from sympy import symbols, Matrix
sympy.init_printing(use_latex="mathjax", fontsize='16pt')
dt = sympy.Symbol('\Delta t')
R = symbols("R")
d = v*dt
a = symbols("a")
alpha = (v + 0.5 * a * dt) * dt / R
omega = symbols("\Omega")
# r = w/sympy.tan(alpha)

R_e = symbols("R_e")

d_r = v * dt / R_e
lat, lng = symbols("lat, lng")
lat_n = sympy.asin(sympy.sin(lat) * sympy.cos(d_r) + sympy.cos(lat) * sympy.sin(d_r) * sympy.cos(theta))
lng_n = lng + sympy.atan2(sympy.sin(theta) * sympy.sin(d_r) * sympy.cos(lat), sympy.cos(d_r) - sympy.sin(lat) * sympy.sin(lat_n))

fxu = Matrix([[lat_n],
              [lng_n],
              [theta+omega*dt],
              [v + a * dt],
              [a],
              [(v + 0.5 * a * dt) / R]])
# fxu = Matrix([[x + v*sympy.sin(theta+alpha) * dt],
#               [y + v*sympy.cos(theta+alpha) * dt],
#               [theta+alpha]])
F = fxu.jacobian(Matrix([lat, lng, theta, v, a, omega]))
F

```
